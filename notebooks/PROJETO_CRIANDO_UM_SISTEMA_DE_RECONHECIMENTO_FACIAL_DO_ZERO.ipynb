```python
import face_recognition
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import files
uploaded = files.upload()
```

```python

image_path = list(uploaded.keys())[0]
print(f"Imagem carregada: {image_path}")
```

```python

image = face_recognition.load_image_file(image_path)
```

```python

face_locations = face_recognition.face_locations(image)
print(f"Faces detectadas: {len(face_locations)}")
```

```python
image_with_boxes = image.copy()

for (top, right, bottom, left) in face_locations:
    cv2.rectangle(image_with_boxes, (left, top), (right, bottom), (0, 255, 0), 2)
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Faces Detectadas")
plt.show()
```

```python
print("Coordenadas das faces detectadas (top, right, bottom, left):")
for i, (top, right, bottom, left) in enumerate(face_locations):
    print(f"Face {i+1}: Top={top}, Right={right}, Bottom={bottom}, Left={left}")
```
